The plt.subplots() cals call allows us to easely create Figure and Axes objects in side by side formation

The plt.subplots() command returns a tuple containing the Fifure canvas and then a numpy array holding the axes objects.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes = plt.subplots(nrows=2, ncols=2)
```

plt.subplots() returns a tuple which by common convention we label(fig, axes):

fig: - This is the entire Figure canvas.

axes: - This is a numpy array holding each of the axes according to position in the overall canvas.

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
fig,axes = plt.subplots(nrows =1,ncols =2)
axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows =1,ncols =2)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows =1,ncols =2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows =2,ncols =2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows =3,ncols = 1)
axes[0].plot(x,y)
axes[2].plot(a,b)

plt.tight_layout()# This Function to arnage these plots.
```

```python
fig,axes = plt.subplots(nrows =2,ncols = 2, figsize = (1,2), dpi =200)
fig.suptitle('Set Of subplots', fontsize=20)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('X-Axis')
axes[0][0].set_ylabel('Y-Axis')
axes[0][0].set_title('First')
axes[0][1].plot(a,b)
axes[0][1].set_xlabel('X-Axis')
axes[0][1].set_ylabel('Y-Axis')
axes[0][1].set_title('Second')
axes[1][0].plot(a,b)
axes[1][0].set_xlabel('X-Axis')
axes[1][0].set_ylabel('Y-Axis')
axes[1][0].set_title('Third')
axes[1][1].plot(x,y)
axes[1][1].set_xlabel('X-Axis')
axes[1][1].set_ylabel('Y-Axis')
axes[1][1].set_title('Fourth')
fig.subplots_adjust(wspace=1, hspace=1)# subplots_adjust(width space, Height space)



fig.savefig('Subplot.png', bbox_inches='tight')
```
