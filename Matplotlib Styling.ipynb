# Matplotlib Styling - Legends

Matplotlib offers very robust styling functions that allows us to edit colors, legends, line widths, markers, and much more

Main Styling Discussed 

legends


Visual Styling: - colors, Editing lines(colors, widths, styles), Editing markers(Colors, Size, Styles, Edges)

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0,11,10)
```

```python
y = x*2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x)
ax.plot(x,y)
ax.plot(x,x**2)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label = 'X vs X')

ax.plot(x,x**2, label = 'X vs X^2')

ax.legend(); # We can pass the parameter in this Function to set the location of legends
# ax.lefend(loc = it may be numeric or may be with name)
```

# Matplotlib Styling 
Visual Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color ='Black',label = 'X vs X'); # We can also give the RGB HEC code for color 

ax.plot(x-4,x*3, color = '#11ed91', label = 'X-4 vs x*3')

ax.legend(loc = 0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color ='Black', lw=10, ls = '-.') # lw means Line Width 

ax.plot(x,x+1, color = '#11ed91',lw=3, ls ='--')# ls means line style

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x, color ='Black', lw=10, ls = '-.') # lw means Line Width 

lines=ax.plot(x,x+1, color = '#11ed91',lw=3)# ls means line style

lines[0].set_dashes([5,2,5,2])
```
